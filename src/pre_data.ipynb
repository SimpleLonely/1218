```python
import numpy as np
import pandas as pd

raw_df = pd.read_csv("../xau_1d_20y.csv")
```

```python
raw_df['log_r'] = np.log(raw_df['close']) - np.log(raw_df['close'].shift(1))

df = pd.DataFrame()
df['date'] = raw_df['date']
df['close'] = raw_df['close']
df['log_r'] = raw_df['log_r'].fillna(0)
df['label'] = np.sign(df['log_r'])
df['label'][df['label']==-1] = 0
```

```python
df.to_csv('../res/input0130.csv')
```
